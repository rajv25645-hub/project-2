```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
a=encoder.fit_transform(df[["Internship_Experience"]]).toarray()
```

```python
encoded_df=pd.DataFrame(a,columns=encoder.get_feature_names_out())
```

```python
final_df=pd.concat([df,encoded_df],axis=1)
```

```python
final_df=final_df.drop(columns=["Internship_Experience"])
final_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder1=LabelEncoder()
```

```python
final_df["Placement"]=encoder1.fit_transform(final_df["Placement"])
```

```python
final_df.head()
```

```python
x=final_df.drop(columns=["Placement"])
```

```python
y=final_df["Placement"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
x=scale.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=56)
```

```python
from sklearn.linear_model import LogisticRegression  
```

```python
regression=LogisticRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
y_pred=regression.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
y_pred=dtc.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knc=KNeighborsClassifier(n_neighbors=15)
```

```python
knc.fit(x_train,y_train)
```

```python
y_pred=knc.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=100)
rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
```
